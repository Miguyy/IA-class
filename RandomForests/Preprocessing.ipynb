# Data Processing

## 1. O que é Data Processing?

De acordo com o Geekforge, pré-processamento de dados é o primeiro passo em qualquer fluxo de trabalho de análise de dados ou de aprendizagem automática. Envolve a limpeza, transformação e organização dos dados brutos para garantir que sejam precisos, consistentes e estejam prontos para a modelação. Tem um grande impacto na construção de modelos, nomeadamente:

- Dados limpos e bem estruturados permitem que os modelos aprendam padrões significativos, em vez de ruído;
- Dados devidamente processados evitam entradas enganosas, levando a previsões mais fiáveis;
- Dados organizados simplificam a criação de entradas úteis para o modelo, melhorando o seu desempenho;
- Dados organizados permitem uma melhor Análise Exploratória de Dados (EDA), tornando os padrões e tendências mais interpretáveis.
