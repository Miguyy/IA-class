# 3. Clear all data

## 3.1. Check for null values

```python
print("Null values in each column: \n", df.isnull().sum())
```

## 3.2. Fill null values with mean / data

```python
df['petal_length (cm)'] = df['petal_length (cm)'].fillna(df['petal_length (cm)'].mean())
df['petal_width (cm)'] = df['petal_width (cm)'].fillna(df['petal_width (cm)'].mean())

print("Null values after filling with mean: \n", df.isnull().sum().sum())
```

# 4. Transformation (Future Engineering)

## 4.1. Converting the column species to numbers 

```python
species_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}

df["target_species"] = df["species"].map(species_mapping)

print("Dataframe with new column 'target_species':")
print(df[['species', 'target_species']].sample(5))
```
